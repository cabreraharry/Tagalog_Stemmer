Next, we will build an array of labels that matches the classification of our articles.  This data type works pretty much like a regular list but is optimized for computations and manipulation.

* `np.zeros()` - create an array of 0's

## `Code-Snippet Exercise #5` (3 points)

Output the following below by applying code below



```python
# select some words to appear in the report. we will assume that each word is unique (i.e. no duplicates)
# WRITE CODE HERE 3 points
keys = ['abot', 'tagal', 'dagdag', 'taas', 'kaso', 'administra', 'lang', 'duterte', 'naman', 'sabi'] 
# list representing our table of word counts.
# each element consist of a sublist with this pattern: [<word>, <positive_count>, <negative_count>]
data = []

# loop through our selected words
for word in keys:
    
    # initialize positive and negative counts
    pos = 0
    neg = 0
    
    # retrieve number of positive counts
    if (word, 1) in freqs:
        pos = freqs[(word, 1)]
        
    # retrieve number of negative counts
    if (word, 0) in freqs:
        neg = freqs[(word, 0)]
        
    # append the word counts to the table
    data.append([word, pos, neg])
    
data
```

We can then use a scatter plot to inspect this table visually. Instead of plotting the raw counts,



```python
fig, ax = plt.subplots(figsize = (8, 8))

# convert positive raw counts to logarithmic scale. we add 1 to avoid log(0)
print(x[2]+1 for x in data)
x = [x[1] + 1 for x in data]

# do the same for the negative counts
y = [x[2] + 1 for x in data]

# Plot a dot for each pair of words
ax.scatter(x, y)  

# assign axis labels
plt.xlabel("Real count Cabrera-HarryEricson-202180224")
plt.ylabel("Fake count Cabrera-HarryEricson-202180224")

# Add the word as the label at the same position as you added the points just before
for i in range(0, len(data)):
    ax.annotate(data[i][0], (x[i], y[i]), fontsize=9)

ax.plot([0, 9], [0, 9], color = 'red') # Plot the red line that divides the 2 areas.
plt.show()
```
